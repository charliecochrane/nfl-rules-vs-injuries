There is a growing body of evidence suggesting that nutrition plays a crucial role in determining life expectancy in various countries. It is widely recognized that poor nutrition can lead to a range of health problems, such as obesity, heart disease, and diabetes, which can significantly reduce life expectancy. On the other hand, a diet that is rich in essential nutrients such as vitamins, minerals, and proteins can promote good health and longevity. Given the importance of nutrition in determining life expectancy, it is reasonable to hypothesize that there is a relationship between a country's nutritional status and its average life expectancy. Therefore, this study aims to investigate the relationship between a country's nutritional status and its life expectancy.

```python
country_life_exp = pd.read_csv("life_expectancy.csv",skiprows=0)
country_diet_info = pd.read_csv("Food_Supply_kcal_Data.csv",skiprows=0,encoding='utf-8')

#Find all the countries in countries' diet DF in the life expectancy DF
country_list = country_diet_info["Country"]
country_life_exp_2 = country_life_exp.loc[country_life_exp["Country"].isin(country_list)]

#Find all the countries in new life expectancy DF in the countries' diet DF
country_list_2 = country_life_exp_2["Country"]
country_diet_info_2 = country_diet_info.loc[country_diet_info["Country"].isin(country_list_2)]

#Checking length
#print(len(country_diet_info_2["Country"]), len(country_list_2))


country_diet_info_2.index = country_diet_info_2['Country']
country_diet_info_2 = country_diet_info_2.sort_index(axis=0,ascending=True) 


country_life_exp_2.index = country_life_exp_2['Country']
country_life_exp_2 = country_life_exp_2.sort_index(axis=0,ascending=True) 

country_diet_info_2["Life Expectancy"] = country_life_exp_2["Life Expectancy   (both sexes)"]

df_country = country_diet_info_2.drop(columns = ["Country", "Life Expectancy", 'Confirmed', 'Deaths',
                                                'Recovered', 'Active', 'Population', 'Unit (all except Population)', 
                                                'Obesity','Undernourished'])
df_country = df_country.dropna()
```

```python
import statsmodels.api as sm
clean_diet_info = country_diet_info_2.dropna()
life_exp = clean_diet_info["Life Expectancy"]
obesity = clean_diet_info["Obesity"]
results = sm.OLS(obesity, life_exp).fit()
print(results.summary())

m = results.params[0]
y = life_exp*m
```

```python
country_diet_info = pd.read_csv("Fat_Supply_Quantity_Data.csv",skiprows=0)
country_diet_info = country_diet_info.drop(columns = ['Confirmed', 'Deaths',
       'Recovered', 'Active', 'Population', 'Unit (all except Population)', 'Obesity', 'Undernourished'])

country_life_exp = pd.read_csv("life_expectancy.csv",skiprows=0)


#Find all the countries in countries' diet DF in the life expectancy DF
country_list = country_diet_info["Country"]
country_life_exp_2 = country_life_exp.loc[country_life_exp["Country"].isin(country_list)]

#Find all the countries in new life expectancy DF in the countries' diet DF
country_list_2 = country_life_exp_2["Country"]
country_diet_info_2 = country_diet_info.loc[country_diet_info["Country"].isin(country_list_2)]

#Checking length
#print(len(country_diet_info_2["Country"]), len(country_list_2))


country_diet_info_2.index = country_diet_info_2['Country']
country_diet_info_2 = country_diet_info_2.sort_index(axis=0,ascending=True) 


country_life_exp_2.index = country_life_exp_2['Country']
country_life_exp_2 = country_life_exp_2.sort_index(axis=0,ascending=True) 

country_diet_info_2["Life Expectancy   (both sexes)"] = country_life_exp_2["Life Expectancy   (both sexes)"]

df_country = country_diet_info_2.drop(columns = ["Country", "Life Expectancy   (both sexes)"])


model_f = sm.OLS(country_diet_info_2["Life Expectancy   (both sexes)"], df_country).fit()
print(model_f.summary())
```

We had three different diet datasets keeping track of the amount of food consumed in units of fat, protein and calories. Aftering fitting a linear model for each dataset we found each model had a high R2 value in the upper 90s, but the P-value for each plot was too high to be considered significant. The plot with the lowest P-value was the dataset recording in calories.



```python
features = ["Sugar Crops", "Animal fats", "Meat"]
for f in features:
    fig = plt.figure( figsize= (20, 15))
    ax = fig.add_subplot(111)
    fig = sm.graphics.plot_regress_exog(model_f, f, fig=fig)
    filename = f + "_Regression_plt" + ".png"
    plt.savefig(filename)
    plt.show()
```

Looking into the Regression plots, we found certain values had decent residual plots with even scatter across the y = 0 intercept. Meat in units of Calories was one of the best plots even though there were a few outliers far below the y = 0. Then there were plots like sugar crops that were more in a vertical single line across y = 0. That means the plot does a bad job fitting that data. Many of the plots for the data points looked like either the sugar crop plot or the meat plot. There were a few in a triangle shape in the residual plot. 




In conclusion, while nutrition is undoubtedly a crucial factor in determining overall health and wellbeing, this study suggests that it may not be the only factor that impacts a country's life expectancy. Although there is some evidence to support the hypothesis that countries with better access to nutritious foods and healthier diets have higher life expectancies, the relationship is not always straightforward.




There are many other factors that may impact a country's life expectancy, including healthcare access, environmental factors, economic stability, and cultural norms. For example, countries with high levels of pollution or limited access to healthcare may experience lower life expectancies despite having a relatively healthy diet. Additionally, cultural norms and traditions may impact diet and health outcomes, making it challenging to draw clear conclusions about the relationship between nutrition and life expectancy.




Therefore, it is important to consider multiple factors when analyzing life expectancy trends in different countries. While nutrition is undoubtedly a critical factor in promoting good health and longevity, it is only one of many factors that impact a country's overall health outcomes.
