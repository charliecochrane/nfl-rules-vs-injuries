```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

food_info = pd.read_csv("Food_Supply_kcal_Data.csv",skiprows=0,encoding='utf-8')
food_info = food_info.drop(columns = ['Confirmed', 'Deaths',
       'Recovered', 'Active', 'Population', 'Unit (all except Population)'])

```

```python
top_countries_l = ['Hong Kong', 'Japan', 'Switzerland', 'Italy', 'Spain', 'Australia', 'Korea, South',
                  'Israel', 'Sweden', 'France','Malta']
top_countries_df = food_info.loc[food_info["Country"].isin(top_countries_l)]
top_countries_df["Top/Bottom Life exp"] = np.ones(len(top_countries_df["Country"]))
top_countries_df
```

```python
bottom_countries_l = ['Central African Republic', 'Chad', 'Lesotho','Nigeria', 'Sierra Leone', 'Cameroon','Mali',
                     'Guinea-Bissau', 'Eswatini', 'Togo']

bottom_countries_df = food_info.loc[food_info["Country"].isin(bottom_countries_l)]
bottom_countries_df["Top/Bottom Life exp"] = np.zeros(len(bottom_countries_df["Country"]))


df_Countries = pd.concat([bottom_countries_df,top_countries_df])
df_Countries.keys()
```

```python
colors = ['red' if label == 0 else 'blue' for label in df_Countries['Top/Bottom Life exp']]

fig = plt.figure(figsize=(15,20))
fig.tight_layout()

plt.subplot(4,2,1)
plt.bar(df_Countries['Country'], df_Countries["Alcoholic Beverages"], color = colors)
plt.rc('axes', titlesize=18) 
plt.title("Alcoholic Beverages")
plt.xticks(rotation = 90)
plt.ylabel("Alcoholic Beveragese consumed (kcal)")
fig.tight_layout()
# plt.savefig('Alcoholic Beveragese(kcal).png')
# plt.show()

plt.subplot(4,2,2)
# fig = plt.figure(figsize=(9,5))
plt.rc('axes', titlesize=18) 
plt.bar(df_Countries['Country'], df_Countries["Meat"], color = colors)
plt.title("Meat")
plt.xticks(rotation = 90)
plt.ylabel("Meat consumed (kcal)")
fig.tight_layout()
# plt.savefig('Meat(kcal).png')
# plt.show()


plt.subplot(4,2,3)
# fig = plt.figure(figsize=(9,5))
plt.rc('axes', titlesize=18) 
plt.bar(df_Countries['Country'], df_Countries["Fruits - Excluding Wine"], color = colors)
plt.title("Fruits - Excluding Wine")
plt.xticks(rotation = 90)
plt.ylabel("Fruits - Excluding Wine consumed (kcal)")
fig.tight_layout()
# plt.savefig('Fruits - Excluding Wine consumed (kcal).png')
# plt.show()



plt.subplot(4,2,5)
# fig = plt.figure(figsize=(9,5))
plt.rc('axes', titlesize=18) 
plt.bar(df_Countries['Country'], df_Countries["Sugar & Sweeteners"], color = colors)
plt.title("Sugar & Sweeteners")
plt.xticks(rotation = 90)
plt.ylabel("Sugar & Sweeteners(kg)")
fig.tight_layout()
# plt.savefig('Sugar & Sweeteners(kcal).png')
# plt.show()

plt.subplot(4,2,4)
# fig = plt.figure(figsize=(9,5))
plt.rc('axes', titlesize=18) 
plt.bar(df_Countries['Country'], df_Countries["Vegetables"], color = colors)
plt.title("Vegetables")
plt.xticks(rotation = 90)
plt.ylabel("Vegetables (kcal)")
fig.tight_layout()
# plt.savefig('Vegetables(kcal).png')
# plt.show()


plt.show()
```
